```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data_df = pd.read_csv("water_potability.csv")
```

```python
data_df
```

```python
from sklearn.impute import KNNImputer
```

```python
imputer = KNNImputer(n_neighbors=5)

imputed_data = imputer.fit_transform(data_df)
```

```python
dd = {
    'ph': imputed_data[:,0],
    'Hardness':imputed_data[:,1],
    'Solids':imputed_data[:,2],
    'Chloramines':imputed_data[:,3], 
    'Sulfate':imputed_data[:,4],
    'Conductivity':imputed_data[:,5],
    'Organic_carbon':imputed_data[:,6],
    'Trihalomethanes':imputed_data[:,7],
    'Turbidity':imputed_data[:,8], 
    'Potability':imputed_data[:,9]
    }

imputed_data_df = pd.DataFrame(data=dd)
```

```python
imputed_data_df
```

```python
import pandas as pd
from ydata_profiling import ProfileReport
```

```python
profile = ProfileReport(imputed_data_df, title="Profiling Report")
```

```python
profile.to_widgets()
```
